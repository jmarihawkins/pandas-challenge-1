## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty'] # create a new column 'subtotal' by multiplying 'unit_price' and 'qty'
display(df[['unit_price', 'qty', 'subtotal']].head()) # this will display the first few rows to verify the new column
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping_price'] = df['unit_weight'].apply(lambda x: 7 if x > 50 else 10) * df['unit_weight'] # calculate shipping price based on unit_weight
display(df[['unit_weight', 'shipping_price']].head()) # this will display the first few rows to verify the new column
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = (df['subtotal'] + df['shipping_price']) * 1.0925 # calculate the total price including sales tax
display(df.head()) # this will display the first few rows to verify the new column
```

```python
# Create a column for the cost of each line using unit cost, qty, and shipping price
# (assume the shipping cost is exactly what is charged to the client)
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price'] # calculate the total cost for each line
display(df.head()) # this will display the first few rows to verify the new column


```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['total_price'] - df['line_cost'] # calculate the profit for each line
display(df[['total_price', 'line_cost', 'line_profit']].head()) # this will print the first few rows to verify the new column

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# Calculate the total spending for each of the top 5 clients
client_spending = df[df['client_id'].isin(top_5_clients)].groupby('client_id')['total_price'].sum()

# Print the total spending for each of the top 5 clients
for client_id in top_5_clients:
    print(f"Client ID {client_id} spent: ${client_spending[client_id]:.2f}")

```

```python
# Create a summary DataFrame showing the totals for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.

# Filter the DataFrame for the top 5 clients
top_5_clients_df = df[df['client_id'].isin(top_5_clients)]

# Group by client_id and calculate the required totals
summary_df = top_5_clients_df.groupby('client_id').agg({
    'qty': 'sum',             # total units purchased
    'shipping_price': 'sum',  # total shipping price
    'total_price': 'sum',     # total revenue
    'line_profit': 'sum'      # total profit
}).reset_index()

# Rename the columns for clarity
summary_df.columns = ['Client ID', 'Total Units Purchased', 'Total Shipping Price', 'Total Revenue', 'Total Profit']

# Convert Total Revenue and Total Profit to formatted string values with $ and commas
summary_df['Total Revenue'] = summary_df['Total Revenue'].apply(lambda x: f"${x:,.2f}")
summary_df['Total Profit'] = summary_df['Total Profit'].apply(lambda x: f"${x:,.2f}")

display(summary_df) # this will print the summary DataFrame


```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns
money_columns = ['Total Revenue', 'Total Profit']

# Define a function that converts a dollar amount to millions
def currency_format_millions(value):
    return f"${value / 1_000_000:.2f}M"

# Apply the currency_format_millions function to only the money columns
summary_df[money_columns] = summary_df[money_columns].replace('[\$,]', '', regex=True).astype(float).applymap(currency_format_millions)

# Rename the columns to reflect the change in the money format
summary_df.columns = [
    'Client ID', 
    'Total Units Purchased', 
    'Total Shipping Price (USD)', 
    'Total Revenue (USD in Millions)', 
    'Total Profit (USD in Millions)'
]

display(summary_df) # this will print the formatted summary DataFrame



```

```python
# Sort the updated data by "Total Profit (USD in Millions)" from highest to lowest and assign the sort to a new DataFrame
sorted_summary_df = summary_df.sort_values(by='Total Profit (USD in Millions)', ascending=False)

display(sorted_summary_df) # this will print the sorted summary DataFrame
```
